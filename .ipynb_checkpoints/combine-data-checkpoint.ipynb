```python
import glob
import pandas as pd
```

```python
# glob.glob('data*.csv') - returns List[str]
# pd.read_csv(f) - returns pd.DataFrame()
# for f in glob.glob() - returns a List[DataFrames]
# pd.concat() - returns one pd.DataFrame()
ridership_data = pd.concat([pd.read_csv(f) for f in glob.glob('initial_data/ridership_data/*.csv')], ignore_index = True)

trip_data = pd.concat([pd.read_csv(f) for f in glob.glob('initial_data/trip_data/*.csv')], ignore_index = True)

```
